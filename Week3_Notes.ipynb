### Histograms


```python
# histogram bar chart shows the freq of a given phenom
fig, ((ax1, ax2,), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = (ax1, ax2, ax3, ax4)

for n in range(0, len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title('n={}'.format(sample_size))
```

### Animation

```python
import matplotlib.animation as animation

n=100
x = np.random.randn(n)
```

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins=bins)
    plt.axis([-4, 4, 0, 30])
    plt.gca().set_title('Sampling the Normal Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n = {}'.format(curr), [3, 27])
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval=100)
```

### Interactivity

```python
# need to be in the artist layer and reference the canvas object of the current figure
# event listing, event driven programming
# piece of data associated with a func call
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at pixels {},{} {}and data {},{}}'.format(event.x,
                                                                        event.y,
                                                                        '/n',
                                                                        event.xdata,
                                                                        event.ydata))  

pl.gcf().canvas.mpl_connect('button_press_event', onclick)
```

```python
# most impt event for us = pick event allows respond when someone actually click on a thing
from random import shuffle
origins = ['China', 'Brazil', 'India', 'USA', 'Canada',
          'UK', 'Germany', 'Iraq', 'Chile', 'Mexico']

shuffle(origins)

df = pd.DataFrame({'height' : np.random.rand(10),
                  'weight' : np.random.rand(10),
                  'origin' : origins})
df
```

plt.figure()

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Selected item came from {}'.format(origin))
plt.gcf().canvas.mpl_connect('pick_event', onpick)
```
