### Plotting with Pandas


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
plt.style.available
```

```python
plt.style('seaborn-colorblind')
```

```python
np.random.seed(123)

    ###seed with the random function holds the numbers product by the random function

df = pd.DataFrame({'a': np.random.randn(365).cumsum(0),
                   'b': np.random.randn(365).cumsum(0) + 20,
                   'c': np.random.randn(365).cumsum(0) - 20},
                   index = pd.date_range('1/1/2017', periods=365))

df.head()
```

```python
df.plot();
```

```python
df.plot('a','b', kind = 'scatter');
```

```python
df.plot.scatter('a', 'c', c='b', s=df['b'], colormap='viridis')
```

```python
### matplotlib subplot returned

ax = df.plot.scatter('a', 'c', c='b', s=df['b'], colormap='viridis')
ax.set_aspect('equal')
```

```python
df.plot.box()
```

```python
df.plot.hist(alpha=0.7);
```

```python
df.plot.kde()

    ###returns a plot the probability density function for the data
    ### come in handy when want to derive a smooth continuous function with a continuous 
    ### dataset
```

### Seaborn


```python
import seaborn as sns
```

```python
np.random.seed(1234) # sets the seed number for the random number generator which lets us reproduce this
#data in downstream calculatoins

v1 = pd.Series(np.random.normal(0,10,1000), name='v1')
v2 = pd.Series(np.random.normal(60,15,1000), name='v2')
plt.hist(v1, alpha=0.7, bins=np.arange(-50, 150, 5), label='v1')
plt.hist(v2, alpha=0.7, bins=np.arange(-50,150,5), label='v2')
plt.legend()
```

```python
plt.figure()
plt.hist([v1, v2], histtype='barstacked', normed=True);
v3 = np.concatenate((v1, v2))
sns.kdeplot(v3);
```

```python
plt.figure()
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'});
```

```python
### joint plot
sns.jointplot(v1, v2, alpha=0.4);
```

```python
grid = sns.jointplot(v1, v2, alpha=0.5);
grid.ax_joint.set_aspect('equal')
```

```python
sns.jointplot(v1, v2, kind='hex');
```

```python
sns.set_style('white')
```

```python
sns.jointplot(v1, v2, kind='kde', space=0);
```
